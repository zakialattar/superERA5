```python
import pandas as pd
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
from utils import * 
import xarray

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import *
from tensorflow.keras import Sequential
from tensorflow.keras import regularizers

import graphviz
import pydot

from keras.utils.vis_utils import plot_model
from keras_visualizer import visualizer 
```

<h2>Prepare Training and Test Sets</h2>

```python
## Define paths to training and test data
cwd = os.getcwd()
train_path = "data/train/"
test_path = "data/test/"

# Training set
X_train_xr = xr.merge([xr.open_dataset(train_path+"inputs_mantua_pt.nc"),xr.open_dataset(train_path+"inputs_staten_pt.nc")])
y_train_xr = xr.merge([xr.open_dataset(train_path+"outputs_mantua_pt.nc"),xr.open_dataset(train_path+"outputs_staten_pt.nc")])
y_train_xr = y_train_xr.assign_coords(lon=X_train_xr.lon.values,lat=X_train_xr.lat.values)


# Test set
X_test_xr = xr.open_dataset(test_path+"inputs_water_pt.nc")
y_test_xr = xr.open_dataset(test_path+"outputs_water_pt.nc")
y_test_xr = y_test_xr.assign_coords(lon=X_test_xr.lon.values,lat=X_test_xr.lat.values)


## Convert to numpy arrays and check to make sure shapes of each dataset match
X_train = X_train_xr[["ssrd","tcc","t2m"]].to_dataframe().drop("expver",axis=1).fillna(0)
y_train_poa = y_train_xr[["time","lat","lon","poa"]].to_dataframe().fillna(0).poa
y_train = pd.DataFrame(X_train.ssrd/3600 - y_train_poa,columns=["res"])

X_test = X_test_xr[["ssrd","tcc","t2m"]].to_dataframe().drop("expver",axis=1).fillna(0)
y_test_poa = y_test_xr[["time","lat","lon","poa"]].to_dataframe().fillna(0).poa
y_test = pd.DataFrame(X_test.ssrd/3600 - y_test_poa,columns=["res"])


print("X_train: " + str(X_train.shape), 
      "\ny_train: " + str(y_train.shape),
      "\nX_test: " + str(X_test.shape),
      "\ny_test: " + str(y_test.shape))
```

```python
x_mean, x_std = X_train.mean(), X_train.std()

X_train_std = (X_train - x_mean) / x_std
X_test_std = (X_test - x_mean) / x_std

y_mean, y_std = y_train.mean(), y_train.std()
y_train_std = (y_train - y_mean) / y_std
y_test_std = (y_test - y_mean) / y_std


print(X_train_std.shape,y_train_std.shape)
print(X_test_std.shape,y_test_std.shape)
```

```python
# set hyperparameters
n_neuron       = 128
activation     = 'relu'
num_epochs     = 200
learning_rate  = 0.01
minibatch_size = 256
model_num      = 1
loss = "mse"

#create the model
model = Sequential([
      keras.layers.Dense(n_neuron, activation=activation, input_shape=(X_train.shape[1],)),
      keras.layers.Dense(n_neuron, activation="exponential"),
      keras.layers.Dense(n_neuron, activation=activation),
      keras.layers.Dense(y_train.shape[1],  activation='linear')
])

model.compile(loss=loss,optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate))

print(model.summary())
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)


early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=40)
print('\nInital test eval on different site:')
model.evaluate(X_test_std, y_test_std)

history = model.fit(X_train_std, y_train_std, 
                    batch_size      = minibatch_size,
                    epochs          = num_epochs,
                    validation_split = .8,
                    verbose         = 1,
                    callbacks       = [early_stop])

print('\nFinal test eval on different site:')
model.evaluate(X_test_std, y_test_std)
```

```python
plot_history(history)
#visualizer(model, format='png', view=True)
```

```python
model_path = os.path.join(cwd,'saved_model')
make_dir(model_path)

# Save the entire model to a HDF5 file.
# The '.h5' extension indicates that the model should be saved to HDF5.
model.save(os.path.join(model_path,'NN_model.h5'))

# reload the saved model
model = load_model(os.path.join(model_path,'NN_model.h5'))

#y_Test 
y_test_predict_std = model.predict(X_test_std)


#plot the results based on residual
df_test = pd.DataFrame()
df_test["y_test_pre_res"] =pd.DataFrame(y_test_predict_std*y_std[0] + y_mean[0]).set_index(X_test.index)[0]
df_test["actual_res"] = y_test_poa - X_test.ssrd/3600

df_plot = df_test[750:1000]
df_plot.plot(figsize=(10,5))
plt.xticks(rotation=45)
df_plot.head(20)
```

```python
#plot the results vs actual temperature
df_test = pd.DataFrame()
df_test["x_test"] = X_test.ssrd/3600
df_test["y_test_pre_comp"] =pd.DataFrame(y_test_predict_std*y_std[0] + y_mean[0]).set_index(df_test.index)[0] + df_test.x_test.values
df_test["actual"] = y_test_poa


df_plot = df_test[750:1000]
df_plot.plot(figsize=(10,5))
plt.xticks(rotation=45)
df_plot.head(20)
```
