```python
from feedinlib import era5
import cdsapi
```

```python
latitude = 39.766039
longitude = -75.142027

for year in ["2021","2022"]:
    for month in range(1,13):
        if month in [1,3,5,7,8,10,12]:
            day = 31
        elif month == 2:
            day = 28
        else:
            day = 30
            
        month = str(month)
        day = str(day)
        # set start and end date (end date will be included
        # in the time period for which data is downloaded)
        start_date, end_date = year+'-'+month+'-01', year+'-'+month+'-'+day
        # set variable set to download
        variable = ["ssrd","tcc","2t"]
        target_file = 'mantua_' +year+'_'+month+'.nc'
        # get windpowerlib data for specified location
        ds = era5.get_era5_data_from_datespan_and_position(
            variable=variable,
            start_date=start_date, end_date=end_date,
            latitude=latitude, longitude=longitude,
            target_file=target_file)
```
